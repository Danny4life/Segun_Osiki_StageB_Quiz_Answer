## Importing all important libriaries

```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

from sklearn.preprocessing import MinMaxScaler

```

```python
df = pd.read_csv('energydata_complete.csv')
```

```python
df.head().T
```

```python
# Drop the date and lights columns
df.drop(['date', 'lights'], axis=1, inplace=True)
```

```python
# Then we normalize our dataset to a normal scale using the MinMaxscaler
scaler = MinMaxScaler()
normalised_df = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
```

```python
# Check the head of our normalized dataset
normalised_df.head().T
```

## Code shown below

```python
# firstly, we select T2 and T6 columns as X and y
X = normalised_df['T2']
y = normalised_df['T6']
```

```python
# We reshape our X column to a two dimensional shape
X = np.array([X]).reshape(-1, 1)
```

```python
# Now, we split our dataset into training and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
model = LinearRegression()
```

```python
# We fit the model to the training dataset
model.fit(X_train, y_train)
```

```python
# Next, we make some predictions
y_pred = model.predict(X_test)
```

```python
# R^2 value
print(f'The R^2 value is:{(r2_score(y_test, y_pred).round(2))}')
```

### Question 17

#### Obtain the feature weights from your linear model above. Which feature has the lowest and highest weights respectively?

#### Answer: RH_2, RH_1

## codes shown below

```python
def feature_weights_df(model_1, feat, col_name):
    '''
    This function gets all the models features weight defined and 
    computes their weights.
    
    Parameters: model_1, feat, col_name.
    '''
    # Define a Series of sort values of all features from lowest to highest
    weights = pd.Series(model_1.coef_, feat.columns).sort_values()
    # Create a DataFrame of the features
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = ['Features', col_name]
    weights_df[col_name].round(3)
    return weights_df
```

```python
linear_model = feature_weights_df(model_1, X_train, 'Linear_Model')
linear_model
```
